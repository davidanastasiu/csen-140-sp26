## Objective
The final step in our data representation journey is handling unstructured text. Because machines cannot 'read' prose, we must map words into a numerical coordinate system. We will explore:
1. **Tokenization & Preprocessing:** Breaking strings into semantic units and removing noise.
2. **The Bag-of-Words (BoW) Model:** Representing documents based on word occurrence.
3. **Sparsity:** Understanding why text data results in 'empty' matrices.
4. **TF-IDF Scaling:** Using math to weight words by their uniqueness and importance.

## 1. Defining the Corpus
We will start with a small dataset where each line represents a document. This sample contains two distinct themes: **Sports** and **Technology**.

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# One document per line
corpus = [
    'The team won the championship game',
    'The quarterback threw a touchdown in the game',
    'The new smartphone has a fast processor',
    'The software update improved the smartphone battery',
    'The coach was happy with the team performance'
]

print(f"Number of documents: {len(corpus)}")
```

## 2. Tokenization and Vocabulary Building
Before encoding, we must clean the text. This involves:
- **Lowercasing:** Ensuring 'Apple' and 'apple' are treated as the same attribute.
- **Stop-word Removal:** Deleting common words (the, a, is) that carry little semantic meaning.
- **Tokenization:** Splitting sentences into individual words (tokens).

```python
# We use CountVectorizer to automate tokenization and stop-word removal
vectorizer = CountVectorizer(stop_words='english')
X_bow = vectorizer.fit_transform(corpus)

# Let's look at our 'Dictionary' (The Attributes)
vocab = vectorizer.get_feature_names_out()
print(f"Vocabulary size: {len(vocab)} unique words")
print(f"Unique words: {vocab}")
```

## 3. The Bag-of-Words Model and Sparsity
In the BoW model, we ignore grammar and word order, keeping only the frequency of words. 

### 3.1 Sparsity
Notice that most documents use only a tiny fraction of the total vocabulary. In a real-world corpus of 100,000 words, a single document might only contain 50 unique words. This leads to a **Sparse Matrix** where nearly all entries are zero.

```python
df_bow = pd.DataFrame(X_bow.toarray(), columns=vocab)

print("Bag-of-Words Matrix (Term Frequency):")
display(df_bow)

# Calculating Sparsity
total_cells = df_bow.size
zero_cells = (df_bow == 0).sum().sum()
sparsity = (zero_cells / total_cells) * 100
print(f"\nMatrix Sparsity: {sparsity:.2f}%")
```

## 4. TF-IDF Scaling
Raw word counts can be misleading. A word like 'smartphone' is very important if it only appears in tech documents, but less helpful if it appeared in every single document in the collection. 

**TF-IDF (Term Frequency-Inverse Document Frequency)** solves this:
- **TF:** Rewards words that appear often in a document.
- **IDF:** Penalizes words that appear too frequently across the entire corpus.

This helps us find **discriminative features**.

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf_vectorizer.fit_transform(corpus)

df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=vocab)

print("TF-IDF Matrix:")
display(df_tfidf.round(3))

# Compare 'game' vs 'touchdown'
# 'game' appears in 2 documents, 'touchdown' appears in 1.
print("\nComparison for Document 1 (Index 1):")
print(f"TF-IDF for 'game': {df_tfidf.iloc[1]['game']:.3f}")
print(f"TF-IDF for 'touchdown': {df_tfidf.iloc[1]['touchdown']:.3f}")
```

## Summary
We have transformed raw text into a high-dimensional vector space. Key takeaways:
1. **Preprocessing** is vital to reduce the dimensionality of our vocabulary.
2. **BoW** captures the essence of a document but results in highly sparse data.
3. **TF-IDF** is a superior representation because it automatically identifies which words are unique to a document's topic.

You are now ready to feed this structured data into Classification or Clustering algorithms!
