## 1. Paradigmatic Similarity in Practice
In lecture, we discussed that word embeddings depend on the notion that similar words occur in similar contexts. This tutorial uses pre-trained GloVe (Global Vectors) to visualize this geometry.

### Learning Objectives:
1. Load real-world word vectors.
2. Visualize clusters of semantically similar words (e.g., 'Biden' and 'President').
3. Perform 'Word Math' (Analogies) using vector arithmetic.

```python
%pip install gensim smart_open

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import gensim.downloader as api

# Load a small pre-trained GloVe model (Wiki 50-dimensional)
print("Downloading GloVe vectors... (approx 66MB)")
model = api.load("glove-wiki-gigaword-50")
```

## 2. Visualizing Word Clusters
Let's look at how words related to 'Government' and 'Food' cluster in separate regions of the embedding space.

```python
words = ["president", "biden", "government", "senate", "pizza", "burger", "restaurant", "chef"]
vectors = np.array([model[w] for w in words])

# Reduce 50D vectors to 2D for plotting
pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

plt.figure(figsize=(10, 8))
plt.scatter(coords[:, 0], coords[:, 1], c='blue')

for i, word in enumerate(words):
    plt.annotate(word, (coords[i, 0], coords[i, 1]), fontsize=12)

plt.title("PCA Projection: Government vs. Food Vectors")
plt.grid(True)
plt.show()
```

## 3. Word Arithmetic (Analogies)
As seen in the lecture, embeddings capture relations like Man:Woman :: King:Queen.
Formula: $King - Man + Woman \approx Queen$

```python
result = model.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)
print(f"King - Man + Woman = {result[0][0]} (Probability: {result[0][1]:.2f})")

result_paris = model.most_similar(positive=['paris', 'germany'], negative=['france'], topn=1)
print(f"Paris - France + Germany = {result_paris[0][0]} (Probability: {result_paris[0][1]:.2f})")
```
