## 1. The Regularization Trade-off
In Support Vector Machines, we balance two competing goals: maximizing the margin and minimizing misclassification. This balance is controlled by two key parameters:

1. **C (Regularization):** 
    * **Large C:** High penalty for misclassification. The model tries to classify every training point correctly, potentially leading to a narrow margin and **High Variance (Overfitting)**.
    * **Small C:** Lower penalty. The model allows some misclassifications to achieve a wider, more robust margin, leading to **High Bias (Underfitting)**.
2. **Gamma (1/Sigma):** In the RBF (Gaussian) kernel, gamma defines how far the influence of a single training example reaches.
    * **Large Gamma (Small Sigma):** Influences are local and "tight." Leads to complex, wiggly boundaries (Overfitting).
    * **Small Gamma (Large Sigma):** Influences are global and "smooth." Leads to simpler boundaries (Underfitting).

**Objective:** We will visualize how these parameters change the decision boundary on a non-linear dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets

# Create a toy non-linear dataset
X, y = datasets.make_moons(n_samples=100, noise=0.15, random_state=42)

def plot_svm_boundary(C, gamma, ax):
    clf = svm.SVC(kernel='rbf', C=C, gamma=gamma)
    clf.fit(X, y)
    
    # Create mesh
    h = .02
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    ax.set_title(f"C={C}, Gamma={gamma}")
```

## 2. Visualizing the Grid
We will now look at four combinations to see the extreme ends of the spectrum.

```python
fig, sub = plt.subplots(2, 2, figsize=(12, 10))
plt.subplots_adjust(wspace=0.4, hspace=0.4)

params = [(0.1, 0.1), (0.1, 10), (100, 0.1), (100, 10)]

for i, (C, gamma) in enumerate(params):
    plot_svm_boundary(C, gamma, sub.flatten()[i])

plt.show()
```

## 3. Analysis
* **Top-Left (0.1, 0.1):** High Bias. The boundary is nearly a straight line and ignores the shape of the data.
* **Top-Right (0.1, 10):** Higher sensitivity to points, but the low C keeps the boundary relatively smooth.
* **Bottom-Left (100, 0.1):** The model is very strict about classification but the low gamma keeps the boundary linear.
* **Bottom-Right (100, 10):** High Variance. The boundary is extremely complex, wrapping tightly around individual points. This model is likely overfit.
